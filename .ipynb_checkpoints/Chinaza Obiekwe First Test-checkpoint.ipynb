### `Question 1`:  What is the difference between 're.findall()' and 're.search()' ? Show example using the Corpus

findall() finds all the matches and returns the matches as a list of strings, with each string representing one match. 
WHILE
re.search() takes a regular expression pattern and a string and searches for that pattern within the string.

They both take two arguments. (patterns, strings).



```python
import re
file = open('fradulent_emails.txt', encoding='utf-8', errors='ignore').read()
# Get Subject
finder = re.search('Subject:.*', file)

# Get Message-Id:
finder2 = re.findall('Message-Id:.*', file)
```

### `Question 3`: Explain the following pattern '\d+\s\w+\s\d+', and what is it used for?


\d+ is usually used to search for dates and \w+ is used to search for names in a text file for example.

\d: matches all numeric characters (0-9)
\+: matches one or more occurences.
Therefore: \d+ matches one or more occurences of numeric characters.

\s: matches all white space characters, and tabs.

\w: matches all alphanumeric characters (A-Z), (a-z), (0-9), dashes(-) and underscores(_)

\w+: matches one or more occurences of alphanumeric characters (A-Z), (a-z), (0-9), dashes(-) and underscores(_) 

These are patterns used in regular expressions. They are used to search and extract data.
For example, an email text file, The (d) is used to identify the date of the message and the plus sign (+) is used to show that the date might come in more than one digit. The (w) is used to get the messages and (s) is used to get the white space characters.
