```python
import requests
from bs4 import BeautifulSoup
import gensim
import docx
url = 'https://www.npr.org/2021/04/28/991683886/frightened-to-death-cheerleader-speech-case-gives-supreme-court-pause'
print("Your url is :", url)
page = requests.get(url).text
print(page)
soup = BeautifulSoup(page)
print(soup)
header = soup.find('h1').get_text()
header2 = 'Summary'
date = soup.find('span', attrs = {'class': 'date'}).text
p_tags = soup.findAll('p')
#strip removes space
text = [tags.get_text().strip() for tags in p_tags]
print(text)
#remove everything thats not new line
sentence_list = [ sentence for sentence in text if not '\n' in sentence]
sentence_list = [ sentence for sentence in sentence_list if '.' in sentence]
article = ''.join(sentence_list)
from gensim.summarization import summarize
summary = summarize(article, ratio = 0.15)
doc = docx.Document()
doc.add_heading(header)
doc.add_picture('picture.png')
doc.add_paragraph(date)
doc.add_paragraph(article)
doc.add_heading(header2)
doc.add_paragraph(summary)
doc.save('document.doc')

```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(summary)
```

### Tokenization

```python
for token in doc:
    print(token.text, '--->', spacy.explain(token.text))
```

### POS Parsing

```python
for token in doc:
    print(token.text, '--->', token.pos_, '--->', spacy.explain(token.pos_))
```

#### Dependency Parsing

```python
for token in doc:
    print(token.text, '--->', token.dep_, '--->', spacy.explain(token.dep))
```

### Lemmatization

```python
for token in doc:
    print(token.text, '--->', token.lemma_, '--->', spacy.explain(token.lemma))
```
