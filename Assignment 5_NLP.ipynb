#### Chinaza Obiekwe. 17100310846

```python
import requests
from bs4 import BeautifulSoup
import gensim
import docx
url = 'https://www.bbc.com/news/world-africa-56993215'
print("Your url is :", url)
page = requests.get(url).text
print(page)
soup = BeautifulSoup(page)
print(soup)
header = soup.find('h1').get_text()
header2 = 'Summary'
date = soup.find('time', attrs = {'data-testid': 'timestamp'}).text
p_tags = soup.findAll('p')
#strip removes space
text = [tags.get_text().strip() for tags in p_tags]
print(text)
#remove everything thats not new line
sentence_list = [ sentence for sentence in text if not '\n' in sentence]
sentence_list = [ sentence for sentence in sentence_list if '.' in sentence]
article = ''.join(sentence_list)
from gensim.summarization import summarize
summary = summarize(article, ratio = 0.15)
doc = docx.Document()
doc.add_heading(header)
doc.add_picture('picture.png')
doc.add_paragraph(date)
doc.add_paragraph(article)
doc.add_heading(header2)
doc.add_paragraph(summary)
doc.save('documentAssignment.doc')

```

import spacy
nlp = spacy.load('en_core_web_sm')

```python
doc = nlp(summary)
```

### Tokenization

```python
for token in doc:
    print(token.text, '--->', spacy.explain(token.text))
```

## POS Parsing

```python
for token in doc:
    print(token.text, '--->', token.pos_, '--->', spacy.explain(token.pos_))
```

## Dependency Parsing

```python
for token in doc:
    print(token.text, '--->', token.dep_, '--->', spacy.explain(token.dep))
```

### Lemmatization

```python
for token in doc:
    print(token.text, '--->', token.lemma_, '--->', spacy.explain(token.lemma))
```

#### Sentence boundary detection

## Named Entity

```python
for ent in doc.ents:
    print(ent.text, '---->', ent.label, '---->', spacy.explain(ent.label))
```

#### Putting it all together

```python
lis = []
```

```python
for token in doc:
    dic = {}
    dic['Token'] = token.text
    dic['POS'] = token.pos_
    dic['Tags'] = token.tag_
    dic['Dep'] = token.dep_
    dic['Explanation'] = spacy.explain(token.tag_)
    lis.append(dic)
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(lis)
```

```python
data
```

```python
l = []
for ent in doc.ents:
    d = {}
    d['entities'] = ent.text
    d['labels'] = ent.label_
    d['Explanation'] = spacy.explain(ent.label_)
    l.append(d)
```

```python
entities = pd.DataFrame(l)
```
